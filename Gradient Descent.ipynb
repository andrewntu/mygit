```python
import math
from math import cos,sin
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import random
from sympy import *
```

## Setting a funcion 
$$f(x)=2sin(x)^3$$

```python
def f(x):
    y = 2*x*np.sin(x)**3
    return y
```

```python
x = np.linspace(-5,5,1000)
initial = random.uniform(-5,5)
y = f(x)
plt.figure()
plt.plot(x,y)
plt.plot(initial,f(initial),'ro')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

## 設定Learning Rate 和 Loss Function

```python
learning_rate1 = 0.01
learning_rate2 = 0.001
learning_rate3 = 1.0
x = Symbol('x')
loss_function = diff(2*x*sin(x)**3,x)
print(loss_function)
```

## 設定迭代次數並找到最小值

```python
def gradient_decsent(itertion,initial_x,learning_rate):
    initialx = initial_x
    initialy = f(initialx)
    histx = []
    histy = []
    for i in range(itertion):
        histx.append(initialx)
        histy.append(initialy)
        initialx = initialx - learning_rate*6*initialx*np.sin(initialx)**2*np.cos(initialx)+ 2*np.sin(initialx)**3
        initialy = f(initialx)
    return histx,histy
result1 = gradient_decsent(20,initial,learning_rate1)
result2 = gradient_decsent(20,initial,learning_rate2)
result3 = gradient_decsent(20,initial,learning_rate3)
```

```python
xp = np.linspace(-5,5,1000)
plt.figure()
plt.plot(result1[0], result1[1],'ro')
plt.plot(result2[0], result2[1],'go')
plt.plot(result3[0], result3[1],'ko')
plt.legend(['Learning Rate = 0.01','Learning Rate = 0.001','Learning Rate = 0.1'])
plt.plot(xp,f(xp))
plt.xlabel('x')
plt.ylabel('f(x)')
```
